#### **Question 7.1:**

**On an average how many games are owned by a user? What is their average playtime? How many reviews did they publish on average?**

While 162 hours seem like a significant amount, when taking the average of products purchased per user (124) into consideration, we discovered that the playtime-to-purchased ratio is 1.31, which means a user spends only about 1.3 hours on each game they own. According to HowLongToBeat, a website that tracks video games completion time, a game’s average length is 18 hours. This indicates that users on Steam tend to purchase more games than can they can complete.

A user on Steam publishes 4 reviews on average. If we consider the average products purchased per user, a user only leaves reviews for 3% of their entire game catalog on the platform. This low number of reviews indicates that there might be higher barriers or no incentives for users on Steam to leave a review. Video game companies may consider creating campaigns to incentivize users to leave positive reviews, since the positive ratio of a game is a big decision factor when considering whether to purchase a game.

```python
%%bigquery
WITH games_owned AS (
  SELECT ROUND(SUM(products) / COUNT(DISTINCT user_id)) AS avg_games_owned_per_user
  FROM (
    SELECT
      user_id,
      products,
      ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY user_id) AS row_num
    FROM `ba-775-project-a04-fall23.SteamGameDataset.rec_users`
  )
  WHERE row_num = 1
),
hours_played AS (
  SELECT ROUND(SUM(hours)/COUNT(DISTINCT user_id)) AS avg_hours_per_user
  FROM `ba-775-project-a04-fall23.SteamGameDataset.rec_users`
),
reviews_published AS (
  SELECT ROUND(SUM(reviews) / COUNT(DISTINCT user_id)) AS avg_reviews_published_per_user
  FROM (
    SELECT
      user_id,
      reviews,
      ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY user_id) AS row_num
    FROM `ba-775-project-a04-fall23.SteamGameDataset.rec_users`
  )
  WHERE row_num = 1
)
SELECT
  COALESCE(games_owned.avg_games_owned_per_user, 0) AS avg_games_owned_per_user,
  COALESCE(hours_played.avg_hours_per_user, 0) AS avg_hours_per_user,
  COALESCE(reviews_published.avg_reviews_published_per_user, 0) AS avg_reviews_published_per_user
FROM games_owned
CROSS JOIN hours_played
CROSS JOIN reviews_published;
```

#### **Question 7.5:**

**Is the average price for Windows-exclusive games higher or lower compared to games that support all operating systems?**

The query results indicate that the average price of a Windows-exclusive game is approximately $5 cheaper compared to games designed to be compatible with all operating systems. One plausible explanation for this difference could be that developing games to be compatible with all operating systems requires more resources. As a result, many game developers may opt to create games exclusively for Windows OS, considering its widespread popularity and a large user base of gamers. This strategic focus on the dominant operating system could contribute to the observed price disparity.

```python
%%bigquery
#The query averages price for games that are windows exclusive and games that are available across all platforms
SELECT
  AVG(CASE WHEN win = true AND mac = false AND linux = false THEN price_final END) AS windows_exclusive_avg_price,
  AVG(CASE WHEN win = true AND mac = true AND linux = true THEN price_final END) AS all_OS_avg_price
FROM `ba-775-project-a04-fall23.SteamGameDataset.g_rec`
```

#### **Question 7.6:**

**What are the top 10 games based on the playtime that have above average rating?**

The outcome of the query shows a trend of highly-rated games on Steam having significant playtime, which is a potential indicator of popularity and engagement among users. The top 10 games with the most hours played, having a rating of ‘Overwhelmingly Positive’,  'Very Positive' or 'Mostly Positive', include titles like 'Lost Ark', 'Mount & Blade II: Bannerlord', and 'Elden Ring’'. We can clearly observe that 'Lost Ark’ game' leads the list with over 27 million hours played and the game with the lowest total playtime among the top 10 games still boasts over 2.6 million hours, indicating a strong engagement even for the titles with the least amount of accumulated playtime in this highly positive-rated segment.

Hours played is a direct metric of engagement, showing how much time players are dedicating to these games.
```python
%%bigquery
SELECT title,
SUM(hours) AS hours, # Calculate the total hours played for each game
MAX(rating) AS rating, # Get the highest rating for each game, in case there are multiple entrie
MAX(discount) AS discount, # Get the highest discount applied to the game
MAX(price_final) AS price_final # Get the highest final price for the game
FROM `ba-775-project-a04-fall23.SteamGameDataset.g_rec`
# Filter to include only games with high user ratings
WHERE rating IN ('Overwhelmingly Positive', 'Mostly Positive', 'Very Positive', 'Positive')
GROUP BY title
ORDER BY hours DESC
Limit 10;
```

#### **Question 7.8:**

**Which top 10 games have the highest total play hours, what are their popularity levels, and what percentage do they represent of the overall playtime?**

Games like "Lost Ark," "Mount & Blade II: Bannerlord," "Call of Duty®," and "ELDEN RING" command significantly high total hours played, indicating a strong player base and prolonged engagement within the users in the Steam community. The games are categorized into popularity levels ('Extremely Popular', 'Very Popular' and 'Moderately Popular') based on their total hours, showing the distinction between games that enjoy exceptionally high playtimes versus those with slightly lower but still considerable engagement.

Gamers who adored one of the extremely popular games "Lost Ark" played 2.7 million hours, had about 129522 active sessions and constituted for 12.68% of total playtime on steam dataset. This is followed by gamers who played "Mount & Blade II: Bannerlord" played 2.5 million hours, had about 127444 sessions and constituted for 11.76% of total playtime on steam dataset. Furthermore, gamers who played "Call of Duty" played 1.5 million hours and had about 114 hours on average, had about 133275 total sessions and constituted to about 7.05% of total playtime on steam dataset.Games like "Lost Ark," "Mount & Blade II: Bannerlord," "Call of Duty®," and "ELDEN RING" command significantly high total hours played, indicating a strong player base and prolonged engagement within the users in the Steam community.

```python
%%bigquery
SELECT
    title,
    SUM(hours) AS total_hours,
    AVG(hours) AS average_hours,
    COUNT(*) AS total_sessions,
    CASE
        WHEN SUM(hours) > 10000000 THEN 'Extremely Popular' #when hours > 10000000, then the game is considered to be extremely popular
        WHEN SUM(hours) > 100000 THEN 'Very Popular' #when hours > 100000, then the game is considered to be very popular
        ELSE 'Moderately Popular' # when hours < 100000, then the game is considered to be moderately popular
    END AS Popularity_Level, # added a new column as popularity level
    (SUM(hours) / (SELECT SUM(hours) FROM `ba-775-project-a04-fall23.SteamGameDataset.g_rec`)) * 100 AS Percentage_Of_Total_Playtime
FROM
    `ba-775-project-a04-fall23.SteamGameDataset.g_rec`
GROUP BY
    title # grouping by title
ORDER BY
    total_hours DESC # filtering by total_hours in descending order
LIMIT 10;
```

####  **Question 7.12:**

**How can we categorize Steam reviewers based on their frequency of reviews, and what insights can be derived by examining the average hours played and average number of games owned within these groups?**

We've categorized Steam users into five groups, ranging from casual to veteran reviewers. Upon examining the average hours played within these groups, an interesting pattern emerged: the overall trend for average hours decreased as we moved towards categories with more reviews. Leveraging our domain knowledge, we formulated a theory suggesting that this phenomenon might be attributed to veteran reviewers receiving complimentary games for review purposes. As a result, they may not fully complete or extensively play these games, leading to a lower average hours played. Additionally, our data confirms that higher-tier categories, especially veteran reviewers, indeed have a significantly larger number of games owned compared to categories with fewer reviews.

```python
%%bigquery
SELECT
    reviewer_category,
    COUNT(user_id) AS count_of_reviewers,
    AVG(hours) AS avg_hours_played,
    AVG(products) AS avg_products_owned
FROM (
    SELECT
        user_id,
        COUNT(review_id) AS review_count,
        AVG(hours) AS hours,
        AVG(products) AS products,
        CASE
            WHEN COUNT(review_id) < 50 THEN 'Casual Reviewers'
            WHEN COUNT(review_id) >= 50 AND COUNT(review_id) < 100 THEN 'Regular Reviewers'
            WHEN COUNT(review_id) >= 100 AND COUNT(review_id) < 200 THEN 'Frequent Reviewers'
            WHEN COUNT(review_id) >= 200 AND COUNT(review_id) < 300 THEN 'Power Reviewers'
            WHEN COUNT(review_id) >= 300 THEN 'Veteran Reviewers'
        END AS reviewer_category
    FROM
        `ba-775-project-a04-fall23.SteamGameDataset.rec_users`
    GROUP BY
        user_id
) AS categorized_reviews
GROUP BY
    reviewer_category;
```

#### **Question 7.14:**


**What is the relationship between game prices and user reviews across different rating categories on Steam in 2022?**

Our analysis of the 2022 Steam dataset revealed insightful trends across rating-based categories of games. We categorized these games into four distinct groups based on their ratings: "Overwhelmingly / Very Positive," "Mostly / Positive," "Mixed / Mostly Negative," and "Negative / Overwhelmingly Negative." This classification aimed to simplify the ratings into broader categories that could help us discern patterns and trends more clearly. Despite "Mixed / Mostly Negative" rated games having the highest average price at USD16.25, they attracted a large number of reviews (711,448), indicating a higher engagement level among users, prompting them to leave feedback despite mixed sentiments. "Mostly / Positive" games, at a lower average price of USD10.58, saw a healthy 518,289 reviews, indicating a good price-to-satisfaction balance. The least expensive games in the "Negative / Overwhelmingly Negative" category had the fewest reviews, only 3,830, at an average price of USD8.07, underscoring limited user engagement. Falling into the "Overwhelmingly / Very Positive" category, had an average price of USD14.42. This group had the highest number of user reviews, reaching an impressive count of 3,512,408. The strong correlation between positive ratings and a high volume of user feedback could indicate a virtuous cycle of quality and engagement. The dataset suggests that while higher-priced games may receive mixed reviews, the highest volume of feedback correlates with highly positive ratings, regardless of price.

```python
%%bigquery
SELECT
  rating_group,
  ROUND(SUM(price_final) / COUNT(DISTINCT app_id), 2) AS average_price_final,
  SUM(user_reviews) AS sum_user_reviews
FROM (
  SELECT
    app_id,
    price_final,
    user_reviews,
    CASE
      WHEN rating IN ('Overwhelmingly Positive', 'Very Positive') THEN 'Overwhelmingly / Very Positive'
      WHEN rating IN ('Mostly Positive', 'Positive') THEN 'Mostly / Positive'
      WHEN rating IN ('Mixed', 'Mostly Negative') THEN 'Mixed / Mostly Negative'
      WHEN rating IN ('Negative', 'Overwhelmingly Negative') THEN 'Negative / Overwhelmingly Negative'
    END as rating_group
  FROM
    ba-775-project-a04-fall23.SteamGameDataset.Games
  WHERE rating IS NOT NULL AND price_final != 0 AND EXTRACT(YEAR FROM date_release) = 2022)
WHERE rating_group IS NOT NULL AND price_final!=0
GROUP BY rating_group
ORDER BY rating_group;
```

#### **Question 7.15:**

**What is the relationship between average playtime, products owned, and user reviews across different user playtime categories on Steam in 2022?**

To see the user behavior of different user segments based on playtime on the Steam platform, we divided all users into four equal groups using playtime, which are heavy users, consistent users, moderate users, and light users. Heavy users exhibited the highest count of total reviews, significantly surpassing other user segments. Interestingly, the intermediary user category displayed a lower count of total reviews in comparison to light users. This observation led to a speculative inference: light users, possessing the highest aggregate of total products owned, might tend to transition swiftly between games, potentially leaving prompt reviews. Conversely, consistent and moderate users, with fewer total games owned, might engage with titles over extended durations, possibly indicating a tendency to refrain from swift judgment or express extreme sentiments, thus influencing their likelihood to contribute reviews.

The segmentation of users based on playtime on the Steam platform revealed distinct patterns in review behavior across user categories. While heavy users displayed significant review contributions, the disparities observed in review counts among user segments underscore potential differences in gaming experiences and reviewing tendencies. The findings suggest a nuanced relationship between playtime, ownership patterns, and review behavior among diverse user cohorts on the platform.

```python
%%bigquery
WITH user_totals AS (
    SELECT
        user_id,
        ROUND(SUM(hours), 2) AS total_hours
    FROM ba-775-project-a04-fall23.SteamGameDataset.rec_users
    GROUP BY user_id
),
max_products AS (
    SELECT
        user_id,
        MAX(products) AS max_product
    FROM ba-775-project-a04-fall23.SteamGameDataset.rec_users
    GROUP BY user_id
),
max_reviews AS (
    SELECT
        user_id,
        MAX(reviews) AS max_review
    FROM ba-775-project-a04-fall23.SteamGameDataset.rec_users
    GROUP BY user_id
),
ranked_users AS (
    SELECT
        ut.user_id,
        ut.total_hours,
        mp.max_product,
        mr.max_review,
        NTILE(4) OVER(ORDER BY ut.total_hours DESC) AS group_number
    FROM user_totals ut
    JOIN max_products mp ON ut.user_id = mp.user_id
    JOIN max_reviews mr ON ut.user_id = mr.user_id
)
SELECT
    CASE
        WHEN group_number = 4 THEN 'Light Users'
        WHEN group_number = 3 THEN 'Moderate Users'
        WHEN group_number = 2 THEN 'Consistent Users'
        WHEN group_number = 1 THEN 'Heavy Users'
    END AS group_name,
    COUNT(user_id) AS users_per_group,
    SUM(total_hours)/COUNT(user_id) AS average_total_playtime,
    MAX(max_product) AS total_products_per_group,
    MAX(max_review) AS total_reviews_per_group
FROM ranked_users
GROUP BY group_number
ORDER BY average_total_playtime DESC;
```

#### **Question 7.17:**


**How have user engagement and reception trends for Steam games developed over time when segmented by the operating system platform, and what is the impact of these trends on the cumulative user reviews and the moving average of positive ratings?**

The results provide a detailed look into Steam games' user engagement and reception for the year 2022, segmented by operating system and the cumulative user reviews with a moving average of positive ratings per game. For instance, the game with the title "Angry Angry Shark," released on 2022-01-01, had a cumulative user review count of 250 at the time of release, and the moving average of positive ratings stood at 75%. These numbers indicate initial user engagement and reception for this specific game.

When examining the broader trends, we'd consider the progression of these numbers across different games and release dates. For example, "Cursed House," released on 2022-01-03, shows a significant increase in cumulative user reviews to 2035 and a moving average of positive ratings at 56.33333%, reflecting changes in user engagement and reception over time. We can compare such figures against the operating system availability (Windows, Mac, Linux) for each title to analyze how platform availability might influence these metrics.

Analyzing these trends involves looking at how the cumulative user review numbers grow over time for each OS and how the moving average of positive ratings fluctuates. This data would help to understand the impact of platform-specific releases on user engagement and game reception over the course of the year.

```python
%%bigquery
WITH MovingAvgData AS (
  SELECT
    title,
    date_release,
    AVG(positive_ratio) OVER (
        ORDER BY date_release
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
    ) AS moving_avg_positive
  FROM `ba-775-project-a04-fall23.SteamGameDataset.Games`
  WHERE EXTRACT(YEAR FROM date_release) = 2022
)
SELECT
  G.title,
  G.date_release,
  G.win,
  G.mac,
  G.linux,
  SUM(G.user_reviews) OVER (
      ORDER BY G.date_release
      ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
  ) AS cumulative_user_reviews,
  M.moving_avg_positive
FROM `ba-775-project-a04-fall23.SteamGameDataset.Games` G
JOIN MovingAvgData M ON G.title = M.title AND G.date_release = M.date_release
WHERE EXTRACT(YEAR FROM G.date_release) = 2022
ORDER BY G.date_release, G.title;
```
